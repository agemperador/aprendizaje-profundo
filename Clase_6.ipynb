```python
import torch.nn.functional as F
import torch
import torchvision as tv
import matplotlib.pyplot as plt
import numpy as np

class RBM (torch.nn.Module):
    
    def __init__ (_,vsize, hsize, CD_K =1): #CD_K es diferencia constractiva, me dice las vueltas que da para entrenar
        
        super().__init__()
        
        _.w = torch.nn.Parameter(torch.randn(hsize,vsize)*1e-2)
        _.bv = torch.nn.Parameter(torch.randn(vsize)*1e-2) #bias visible
        _.bh = torch.nn.Parameter(torch.randn(hsize)*1e-2)
        _.k = CD_K
        
    def sample_h (_,v): #Para estimar las ocultas le pasamos las visibles _.v
        
        prob_h = torch.sigmoid(F.linear(v,_.w,_.bh)) #Estimulo v por W y restamos bias y sigmoid
        samp_h = torch.bernoulli(prob_h) # Devuelve un vector con la misma dimension pero con valores 0 y 1
        return prob_h, samp_h
        
    def sample_v(_,h):

        prob_v = torch.sigmoid(F.linear(h,_.w.t(),_.bv)) #Estimulo v por W y restamos bias y sigmoid
        samp_v = torch.bernoulli(prob_v) # Devuelve un vector con la misma dimension pero con valores 0 y 1
        return prob_v, samp_v
        
    def free_energy (_,v): ## Defino mi loss function
        v_bv = v.mv(_.bv) ## Multiplica matriz por vector
        hlin = torch.clamp(F.linear(v,_.w, _.bh),-80,80) ## Clamp evita que me de valores muy grandes para que no diverja
        slog = hlin.exp().add(1).log().sum(1)
        return (-slog - v_bv).mean()
    
    def forward(_,v):
        vs = v
        for i in range(_.k): #Iteramos k veces (k Vueltas)
            hp, hs = _.sample_h(vs)
            vp, vs = _.sample_v(hs)
        return v,vs
```

## Para poder importar una función de un .py y que no se ejecute nada mas que la función, todo lo que sea ejecutable tiene que estar dentro de if __name__ == '__main__' 

## Es lo que verifica que el código se esta ejecutando desde el mismo archivo y no de otro

```python
if __name__ == '__main__':
    
    T = 20
    B = 100 # Tamaño del minilote
    
    """""""""""""""""""""""""""""""""""""""""""""TRAIN Y TEST """""""""""""""""""""""""""""""""""""""""""""
    # Cargo un directorio ./data y baja los datos de entrenamiento - totensor los convierte en tensor
    trn_data = tv.datasets.MNIST(root='./data', train = True,  download = True, transform = tv.transforms.ToTensor() )
    # Lo mismo para el test data
    tsn_data = tv.datasets.MNIST(root='./data', train = False, download = True, transform = tv.transforms.ToTensor() )
    # Se puede trabajar automaticamente con mini batch
    trn_load = torch.utils.data.DataLoader (dataset = trn_data, batch_size = B, shuffle = True)
    tsn_load = torch.utils.data.DataLoader (dataset = tsn_data, batch_size = B, shuffle = True)

    """"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
    
    N = 784
    M = 64
    C = 10
    
    #Queremos detectar características, Restricted Boltzmann Machine Feature Detection
    rbmfd = RBM(N,M)
    optim = torch.optim.SGD(rbmfd.parameters(),0.1)

    rbmfd.train()
    print('Máquina de Boltz:')
    for t in range(T):
        err = []
        for images, labels in trn_load:
        
            optim.zero_grad()


            data = images.view(-1,N)
            v0, vk =  rbmfd(data)
            loss = rbmfd.free_energy(v0) - rbmfd.free_energy(vk)
            
            loss.backward()
            
            err.append(loss.item())

            optim.step()
        print(t, 'de', T)
        #plt.plot(err)
        #plt.show()
    
    lincl = torch.nn.Linear(M,C)
    optim = torch.optim.SGD(lincl.parameters(),0.1)

    lostf = torch.nn.CrossEntropyLoss()
    print('Última capa de clasificación')
    for t in range(T):
        err = []
        for images,labels in trn_load:
            optim.zero_grad()
            
            v = images.view(-1,N)
            hp,hs = rbmfd.sample_h(v)
            cp = lincl(hp)
            
            loss = lostf(cp,labels)

            loss.backward()
            err.append(loss.item())
            optim.step()
        print(t,'de', T)
        plt.plot(err)
        plt.show()
        
        
```

```python

lincl = torch.nn.Linear(M,C)
optim = torch.optim.SGD(lincl.parameters(),0.1)

lostf = torch.nn.CrossEntropyLoss()
print('Última capa de clasificación')
for t in range(T):
    err = []
    for images,labels in trn_load:
        optim.zero_grad()

        v = images.view(-1,N)
        hp,hs = rbmfd.sample_h(v)
        cp = lincl(hp)

        loss = lostf(cp,labels)

        loss.backward()
        err.append(loss.item())
        optim.step()
    print(t,'de', T)
    plt.plot(err)
    plt.show()

```

```python
"""""""""TEST"""""""""
lincl.eval()

#HOLD OUT es lo mas simple que se puede hacer


right, total = 0,0

with torch.no_grad():
    for images, labels in  tsn_load:

        v = images.view(-1,N)
        
        hp,hs = rbmfd.sample_h(v)

        y = lincl(hp)
        
        right += (y.argmax(dim = 1) == labels).sum().item()
        total += len(labels)
    accuracy = right / total
#rbmfd.parameters
print(accuracy,right,total)

```
