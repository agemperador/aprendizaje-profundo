### Esta clase implementamos una DBN Deep Belief Network
### Osea, apilamos RBM y al final usamos un clasificador lineal.

```python
import torch, torchvision as tv

import matplotlib.pyplot as plt
from Boltzman import RBM

```

```python
Modelo de hinton [28x28, 500, 500, 2000, 10]
```

```python
class DBN (torch.nn.Module):
    
    def __init__(_,sizes, CD_K=1): #Divergencia constructiva
        
        super().__init__()
        _.subnet = torch.nn.ModuleList()
        
        # hinton [28x28, 500, 500, 2000, 10] 
        # recorro hasta -2 para pararme en 500 y que la ultima capa sea de  [500,2000]
        # El 10 es para el clasificador lineal
        for i in range(len(sizes)-2):
            _.subnet.append(RBM(sizes[i],sizes[i+1],CD_K))
            
        #Clasificador lineal
        _.output = torch. nn.Linear(sizes[-2], sizes[-1])
        
    def forward(_,v, depth = None): # deptl es para decirle que capa quiero entrenar, si es None lo toma como toda la red
        vi = v
        if depth is not None:
            for i in range(depth):
                hp,vi = _.subnet[i].sample_h(vi) # Lleva la entrada a la capa indicada
            vp,vo =  _.subnet[depth].forward(vi)
        else: 
            for rbm in _.subnet:
                hp,vi = rbm.sample_h(vi)
            vo = _.output(hp) #Modelo lineal
            
        
        return vi,vo
            
        
        
```
