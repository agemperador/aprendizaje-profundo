## Defino la clase convolucional

Hay que usar ReLu porque es mas específica.
Muchas unidades dan 0 y unas pocas dan positivas.

Si usamos sigmoide es muy borroso y continuo, no se nota bien.

La gpu es muy rápida para hacer algunas cosas y otras no.
No hay tgh ni sigmoid en gpu, pero si max()


```python
class Convnet(torch.nn.Module):
    
    def __init__(_,k1=5,s1=2,p1=2,k2=3,s2=1, p2=1, k3=2, s3=2):
        super().__init__()
        
        
        _.c1 = torch.nn.Conv2d(in_channels=3, out_channels=16, kernel_size=k1, stride= s1, padding=p1)
        _.c2 = torch.nn.Conv2d(in_channels= 16, out_channels= 32, kernel_size= k2, stride=s2, padding=p2)
        _.mp = torch.nn.MaxPool2d(kernel_size=k3, stride= s3)
        _.l1 = torch.nn.Linear(32*8*8,512)
        _.l2 = torch.nn.Linear(512,10)
        
    def forward(_,x):
        
        h1 = _.c1(x)
        h2 = _.c2(h1)
        h3 = _.mp(h2).relu()
        h4 = _.l1(h3.view(-1,32*8*8)).tanh()
        h5 = _.l2(h4)
        return h5
        
        
        
        
        
        
```

```python
"""idata = iter(trn_load)
image , label = next(idata)
print(image.size(), image.shape)
plt.imshow(image[0,0,...])
plt.show()"""

model = Convnet()

optim = torch.optim.SGD(model.parameters(),0.01)
lostf = torch.nn.CrossEntropyLoss()

model.train()


T=10
for t in range (T):

    err = []
    er = 0
    for images,labels in trn_load:
        optim.zero_grad() 
        y = model(images)
        loss = lostf(y,labels)

        loss.backward()
        err.append(loss.item())


        optim.step()

    print(t,'de', T)
    print(torch.mean(err))
    plt.plot(err)
    plt.show()

```

```python
"""""""""TEST"""""""""
model.eval()

#HOLD OUT es lo mas simple que se puede hacer


right, total = 0,0

with torch.no_grad():
    for images, labels in  tsn_load:

        #plt.imshow(images[0,0,:,:])
        #plt.show()
        y = model(images)
        #print (y[1])
        right += (y.argmax(dim = 1) == labels).sum().item()
        total += len(labels)
    accuracy = right / total
#rbmfd.parameters
print(accuracy*100,right,total)
```

```python

class ConvnetAE(torch.nn.Module):
    
    def __init__(_,k1=5,s1=2,p1=2,k2=3,s2=1, p2=1, k3=2, s3=2):
        super().__init__()
        
        _.enc = torch.nn.Sequential(
                        torch.nn.Conv2d(in_channels=3, out_channels=16, kernel_size=k1, stride= s1, padding=p1),
                        torch.nn.Conv2d(in_channels= 16, out_channels= 32, kernel_size= k2, stride=s2, padding=p2),
                        torch.nn.Conv2d(in_channels= 32, out_channels= 32,kernel_size=k3, stride= s3),
                        torch.nn.ReLU(True)  
                        )
        _.l1 = torch.nn.Linear(32*8*8,512)
        _.l2 = torch.nn.Linear(512,10)
        _.dl1 = torch.nn.Linear(10,512)
        _.dl2 = torch.nn.Linear(512,32*8*8)
        _.dec = torch.nn.Sequential(
                        torch.nn.ConvTranspose2d(in_channels= 32, out_channels= 32,kernel_size=k3, stride= s3),
                        torch.nn.ConvTranspose2d(in_channels= 32, out_channels= 16, kernel_size= k2, stride=s2, padding=p2),
                        torch.nn.ConvTranspose2d(in_channels=16, out_channels=3, kernel_size=k1, stride= s1, padding=p1),
                        )
        
        
    def forward(_,x):
        
        enc = _.enc(x)
        #print(enc.shape)
        out1 = _.l1(enc.view(-1,32*8*8)).tanh()
        #print(out1.shape)
        out2 = _.l2(out1).tanh()
        #print(out2.shape)
        
        h = _.dl1(out2).tanh()
        #print(h.shape)
        h1 = _.dl2(h).view(-1,32,8,8).tanh()
        #print(h1.shape)
        dec = _.dec(h1)
        #print(dec.shape)
        return dec.tanh()
    def generate (_,x):
    
        h = _.dl1(x)
        h1 = _.dl2(h).view(-1,8,8)
        print(h1.shape)
        dec = _.dec(h1)
        
        return dec
        
    
idata = iter(trn_load)
image , label = next(idata)   
"""
model = ConvnetAE()
y = model(image[...])
plt.imshow(y[0,0,:,:].detach().numpy())"""
```

```python
model = ConvnetAE()
optim = torch.optim.SGD(model.parameters(),0.01)
lostf = torch.nn.MSELoss()

model.train()
T= 10

for t in range(T):
    err = []
    for images , labels in trn_load:
        
        optim.zero_grad()
        
        data = images
        y = model(data)
        
        loss = lostf(y,data[:,:,:-1,:-1])
        
        loss.backward()
        err.append(loss)
        optim.step()
    print('%i de %i'%(t,T))
plt.plot(err)
plt.show()
```
