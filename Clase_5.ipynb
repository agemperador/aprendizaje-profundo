```python
import torch
import torchvision as tv
import matplotlib.pyplot as plt
import numpy as np

# Cargo un directorio ./data y baja los datos de entrenamiento - totensor los convierte en tensor
trn_data = tv.datasets.MNIST(root='./data', train = True,  download = True, transform = tv.transforms.ToTensor() )

# Lo mismo para el test data
tsn_data = tv.datasets.MNIST(root='./data', train = False, download = True, transform = tv.transforms.ToTensor() )

# Tamaño del mini lote
B = 100

# Se puede trabajar automaticamente con mini batch
trn_load = torch.utils.data.DataLoader (dataset = trn_data, batch_size = B, shuffle = True)
tsn_load = torch.utils.data.DataLoader (dataset = tsn_data, batch_size = B, shuffle = True)

P = len(trn_data)
N = trn_data[0][0].nelement()
C = 10


class mlp(torch.nn.Module): #Multi Layer perceptron
    
    def __init__(_,isize,hsize,osize ): #self o _ para llamarse a si mismo, unidades de entrada, unidades ocultas, unidades de salida
        
        #para que llame al modulo torch.nn.Module y haga las operaciones que tenga que hacer, se usa super
        super().__init__()
        
        #Defino las capas
        _.L1 = torch.nn.Linear(isize, hsize)
        _.L2 = torch.nn.Linear(hsize, osize)
        
    def forward(_, x):
    
        h = torch.tanh(_.L1(x))
        y = torch.softmax(_.L2(h),dim = 1)
        
        return y

# Mas capas para mas placer
    
class MLP (torch.nn.Module):
    
    #sizes incluye la entrada y la salida
    def __init__ (_,sizes):
        
        super().__init__()
        
        _.layers = torch.nn.ModuleList()
        
        for i in range(len(sizes)-1):
            _.layers.append(torch.nn.Linear(sizes[i],sizes[i+1]))
        
    def forward(_,x):
        
        h = x
        # Itero todas las capas menos la ultima
        for hidden in _.layers[:-1]:
            h = torch.sigmoid (hidden (h))
        #La ultima es output
        output = _.layers[-1]
        #La uso con softmax
        y = output(h)
        
        #No uso softmax porque uso crossentropy
        return y
       
```
